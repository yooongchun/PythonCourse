## 数据持久化

```python
import pickle

d={'course':'Python基础课程','student':'零基础','feature':'有趣有用'}
print(type(d))
print(d)
# 保存
with open('data.dict','wb')as fp:
    pickle.dump(d,fp)

# 加载
with open('data.dict','rb')as fp:
    d2=pickle.load(fp)
print('d2==d:',d2==d)
print(d2)
```

## 多线程

```python
import threading
import time
from datetime import datetime as dt
def say_hello(word):
    time.sleep(1)
    print(word)

pool=[]
now=dt.now()
for x in range(10):
    t=threading.Thread(target=say_hello,args=(x,))
    t.start()
    pool.append(t)
for t in pool:
    t.join()
end=dt.now()
print('总用时：',(end-now).total_seconds())
```
